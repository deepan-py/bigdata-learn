```python
%%sql

SELECT department_name, employee_id, first_name, last_name, salary FROM employees e1 LEFT JOIN departments USING (department_id) 
WHERE salary = (SELECT max(salary) FROM employees e2 WHERE e1.department_id=e2.department_id GROUP BY department_id) ORDER by department_name;
```

```python
%%sql 

SELECT department_name, first_name, last_name, t.salary from departments d 
inner join (SELECT department_id, max(salary) as salary FROM employees GROUP by department_id) t on t.department_id=d.department_id 
inner JOIN employees e on e.department_id=t.department_id and e.salary=t.salary order by department_name
```
